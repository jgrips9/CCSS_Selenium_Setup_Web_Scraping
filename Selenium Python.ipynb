```python
#Start web driver
chrome_options = webdriver.ChromeOptions()
driver = webdriver.Chrome(options=chrome_options)
```

```python
driver.get("https://statements.cornell.edu/pollack.cfm")
```

```python
articles = driver.find_elements(By.XPATH, "//a[@class='cu-statement-title']")
```
